```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
N = 1000

std = 1.5

bad_width = np.random.normal(10, std, N)
bad_height = np.random.normal(30, std, N)
bad_weight = np.random.normal(0.35, 0.3, N)

kh_width = np.random.normal(30, std, N)
kh_height = np.random.normal(10, std, N)
kh_weight = np.random.normal(7, std, N)

width = np.concatenate((bad_width, kh_width))
height = np.concatenate((bad_height, kh_height))
weight = np.concatenate((bad_weight, kh_weight))

X = np.array([width, height, weight]).T

bad_label = np.zeros(N, dtype='int')
kh_label = np.ones(N, dtype='int')

Y = np.concatenate((bad_label, kh_label))

ax.scatter(X[0:N, 0], X[0:N, 1], X[0:N, 2], c='orange')
ax.scatter(X[N:2*N, 0], X[N:2*N, 1], X[N:2*N, 2], c='red')

ax.set_xlabel('Width')
ax.set_ylabel('Height')
ax.set_zlabel('Weight')
plt.title("Persian Melon and Balloon")
```
